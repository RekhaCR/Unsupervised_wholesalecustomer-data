```python
#Outlier detection, measure in percentage
num_col = df.columns.tolist()
#Function to detect the outliers using IQR
def outlier_count(col, data=df):
    #q75, q25 = np.percentile(data[col], [25, 75])
    # calculate the interquartile range(Q1,Q3)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    min_val = Q1 - (IQR*1.5)
    max_val = Q3 + (IQR*1.5)
    #Finding the length of data that is more than max threshold and lesser than min threshold
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    print('{:<20} {:<20} {:.2f}%'.format(col,outlier_count,outlier_percent))
    
#Looping over all the numerical columns to outlier count function to find the total count of outliers in data.
print("\n"+20*'*' + ' Outliers ' + 20*'*'+"\n")
print('{:<20} {:<20} {:<20}'.format('Variable Name','Number Of Outlier','Outlier(%)'))
for col in num_col:
    outlier_count(col)
```

# Clustering without using PCA

KMeans Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)    #instance is ready
    kmeans.fit(data_scaled)                                                   #fitting the model
    wcss.append(kmeans.inertia_)
print(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans= kmeans.fit_predict(data_scaled)
print(y_kmeans)
```

### We can use Silhouette score ito evaluate the quality of clusters created using clustering algorithms such as K-Means in terms of how well samples are clustered with other samples that are similar to each other. 

```python
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=5): {silhouette_score(data_scaled, y_kmeans)}')
```

```python
kmeans = KMeans(n_clusters =4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(data_scaled)
y_kmeans
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg=silhouette_score(data_scaled,y_kmeans)
print(round(silhouette_avg,4))
```

By examining through different number of samples the optimum number of samples which gave better score was found to be 4

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
plt.figure(figsize=(12,10))
dendrogram(linkage(data_scaled, method = 'ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

In the dendrogram locate the largest vertical difference between nodes, and in the middle pass an horizontal line. The number of vertical lines intersecting it is the optimal number of clusters and here it is 2.

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity ='euclidean', linkage ='ward')
y_hc = hc.fit_predict(data_scaled)
y_hc
```

```python
silhouette_avg=silhouette_score(data_scaled,y_hc)
print(round(silhouette_avg,5))
```

### The number of optimul clusters for kmeans is 4 and the optimum number of clusters in heairachial clustering is found to be 2

# Clustering using PCA

Principal component analysis, or PCA, is a statistical procedure that allows you to summarize the information content in large data tables by means of a smaller set of “summary indices” that can be more easily visualized and analyzed. Here we are going to check if the dimensionality reduction can improve the clustering scores of the optimized kmeans and heirarchial clustering models done in the above section.

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_scaled =sc.fit_transform(X)
data_scaled=pd.DataFrame(data_scaled)
data_scaled
```

## PERFORMING PCA ON KMEANS ALGORITHM

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
pca.fit(data_scaled)
x_pca=pca.transform(data_scaled)
x_pca.shape
```

```python
x_pca
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)    #instance is ready
    kmeans.fit(x_pca)                                                   #fitting the model
    wcss.append(kmeans.inertia_)
print(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show
```

```python
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=42)
y_kmeans= kmeans.fit_predict(x_pca)
print(y_kmeans)
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=5): {silhouette_score(x_pca, y_kmeans)}')
```

```python
pca=PCA(n_components=2)
pca.fit(data_scaled)
x_pca=pca.transform(data_scaled)
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=42)
y_kmeans= kmeans.fit_predict(x_pca)
print(y_kmeans)
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=5): {silhouette_score(x_pca, y_kmeans)}')
```

```python
plt.figure(figsize=(12,10))
dendrogram(linkage(x_pca, method = 'ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

Optimum number can be selected as 2

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity ='euclidean', linkage ='ward')
y_hc = hc.fit_predict(x_pca)
y_hc
```
